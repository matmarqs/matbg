```python
import numpy as np
from numpy.linalg import eigh
from numpy.linalg import norm
import matplotlib.pyplot as plt
```

```python
a = 0.246  # in nanometers
th = 1.05 * np.pi/180
kD = 4*np.pi/3*a
kth = kD * 2 * np.sin(th/2)
qb  = kth * np.array([0, -1])
qtr = kth * np.array([np.sqrt(3)/2, 1/2])
qtl = kth * np.array([-np.sqrt(3)/2, 1/2])
q = np.array([qb,qtr,qtl])
b1 = q[1]-q[2]
b2 = q[0]-q[2]
b3 = q[1]-q[0]
b = np.array([b1,b2,b3])
cut = 4
vf = 1
u = 0.11
up = 0.11
```

```python
Q = np.array([np.array(list([i,j,0]@b - l*q[0]) + [l]) for i in range(-100,100) for j in range(-100,100) for l in [0,1] if norm([i,j,0]@b - l*q[0]) <= np.sqrt(3)*kth*cut])
Nq = len(Q)
print(Nq)
```

```python
Q_nn = {}
for i in range(Nq):
	Q_nn[i] = [[np.round(Q[:,:2],3).tolist().index(list(np.round(Q[i,:2]+q[j],3))),j] for j in range(len(q)) if list(np.round(Q[i,:2]+q[j],3)) in np.round(Q[:,:2],3).tolist()]
```
